## 特徴量エンジニアリング
- すでにある特徴量から、予測に有益な情報を作り出す
- 非常に重要な処理（前処理、特徴量選択、ハイパーパラメーターチューニングよりも有効）
- どんなに良いモデルや最適なハイパーパラメータを選択しても良い特徴量がないと高い精度は出せない。非常に重要な操作
- なるべく多く作って試すのがよい
    - 日付情報からの年、月、日、曜日等
    - 多項式特徴量
    - binning(数値ー＞カテゴリカル)
    - 四則演算（ex,　利益　＝　売上 - 経費）
    - 他テーブルからの積み上げ値
    - ドメイン的に意味のある特徴量（submodelの予測値から作成した特徴量）
    
## 日付情報からの年、月、日、曜日等
- 最もよく使われる特徴量エンジニアリング
- 時系列データに限らず有効
- データの性質によっては注意が必要な場合もある
    -　　例：テストデータが学習データの後続データである場合はYear情報をそのまま使えない
    - 例：2年未満のデータの場合、月の周期的な季節性を捉えられない
- 時、分、秒、etcはデータの粒度に応じて分析・処理をする
    - Date: 2014/5/1 {day of week: 3(tuesday) / day of year: 121 / quater: 2 / week of year: 18 / leap year: False}
    
## 多項式特徴量(polynomial features)
- 特徴量を多項式化することで非線形化する
- 元の特徴量空間では非線形になるが、変換後の特徴量の特徴量空間では線形モデルのまま
#### 交互作用項
- 特徴量が複数ある場合は交互作用項を作ることもできる
- 交互作用項を使う場合はそれぞれの特徴量も使うのが一般的
- 全ての交互作用項や多項式項を使う必要はない
    - 例：「広さ」と「駅からの距離」の例ー＞　広くて駅から近いと相乗効果でさらに家賃が上がる
#### 「補足」決定木と交互作用項
- 決定木は分岐を繰り返すことで交互作用項を表現できる
- 決定木（GBDT含む）を使う場合は無理に色々な種類の交互作用項を作る必要がない

## binning
- 数値の特徴量からカテゴリカルの特徴量を生成する
- 元のカラムと組み合わせで使用してもOK
- 実際には全データの範囲に対してパーセンタイルによってbinningすることが多い

## 四則演算
- 特徴量同士を四則演算して新たな特徴量を生成する
- 網羅的にやるのは大変なので、ドメイン的に意味のある特徴量を生成する
- 集約値との四則演算
    - 例："good" + "east" = "good_east"
    - target encoding でより目的変数とカテゴリ値の組み合わせを細分化できる
        - 過学習には注意が必要（カテゴリに対してのレコード数が減るので）
    - 例：　利益 = 売上 - 経費
    - 商品のカテゴリ内の売上比率 = 当該商品の売上　/ 当該カテゴリの売上合計
## 集約値
- 1:Nの関係の2つのテーブルがある場合、テーブル結合のサイン集約値を使う
- min, max, sum, mean, median, std... etc

## submodelを使った予測値
- 本来の目的変数以外の特徴量をtargetにしてモデルを学習する
- 欠損値補完以外にも使うことができる
- ドメイン知識的に意味のある変数を作ることで、精度の良いモデルを作成することが可能

# 日付

```python
import pandas as pd
df = pd.read_csv('bike_share.csv')
```

```python
df
```

```python
# 日付カラムの確認
df.info()
```

```python
type(df['datetime'].iloc[0])
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.info()
```

```python
# 日付データからそれぞれの値を取得 dtというオブジェクトに対して欲しいものを指定する

df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['dayofyear'] = df['datetime'].dt.dayofyear
df['dayofweek'] = df['datetime'].dt.dayofweek
df['weekofyear'] = df['datetime'].dt.isocalendar().week
df['quarter'] = df['datetime'].dt.quarter
df['isleap'] = df['datetime'].dt.is_leap_year
```

```python
# 新しい特徴量の確認　
# 1/3　月曜日なので'dayofweek'がリセットされる。dayofweekは月曜日がスタートである
# 結構簡単に作れるのでおすすめ

df
```

### 多項式特徴量

```python
df = pd.read_csv('penguins_size.csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import PolynomialFeatures
import seaborn as sns

# degreeは何乗するか。　普通はバイアス項はいらないのでinclude_biasはfalseに指定

poly = PolynomialFeatures(degree=2, include_bias=False)
df[['culmen_length_mm', 'culmen_depth_mm','culmen_length_mm^2', 'culmen_length_mm * culmen_depth_mm', 'culmen_depth_mm^2' ]] = poly.fit_transform(df[['culmen_length_mm', 'culmen_depth_mm']])
df
```

```python
sns.pairplot(df, hue='species')
```

# binning

```python
# 数値の特徴量からbinningしてカテゴリカル変数を作る
df['body_mass_g_bin'] = pd.cut(df['body_mass_g'], bins=10, labels=False)

# カテゴリカル変数として扱う場合dtypeをobjectに変換．数値変数として扱ってもよい
df['body_mass_g_bin'] = df['body_mass_g_bin'].astype('object')
df.info()
```

# 四則演算

```python

df['culmen_diff'] = df['culmen_length_mm'] - df['culmen_depth_mm']
df['culmen_ratio'] = df['culmen_length_mm'] / df['culmen_depth_mm']
%matplotlib inline
sns.pairplot(df, hue='species')

# 'clumen_ratio'は分布を見ると綺麗に分かれているように見えます。なので'clumen_ratio'はかなり強力な特徴量になり得ます
```

# 集約値

```python
# SK_ID_CURRをキーに過去の累計借用額(credit)の合計を特徴量にする
# AMT_creditが過去の借用額

df_app = pd.read_csv('homecredit_application.csv')
df_prev_app = pd.read_csv('homecredit_previous_application.csv')
```

```python
df_app.head(3)
```

```python
df_prev_app.sort_values('SK_ID_CURR').head()
```

```python
# 集約値計算 'SK_ID_CURR'でgroupbyし、'AMT_CREDIT'を合計する
credit_prev_sum = df_prev_app.groupby('SK_ID_CURR').sum()[['AMT_CREDIT']]
credit_prev_sum
```

```python
# SK_ID_CURRをキーに表を結合
df_app.join(credit_prev_sum, on='SK_ID_CURR', rsuffix='_PREV_SUM')
```
