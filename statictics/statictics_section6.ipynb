## 統計的仮説検定  
・統計的に仮説が正しいかを確かめる  
・意思決定を左右する重要な役割を果たす  
・多くの検定がある。例）２群の比率差の検定、連関の検定、２群の平均値差の検定、２群の等分散の検定、正規性の検定  
例）DMの施策を変更したらメールからのWeb訪問数が増えたか？薬Aと薬Bよりも高価が高いと言えるか。偶然ではないことを言いたい。  


## 帰無仮説（null hypothesis)と対立仮説（alternative hypothesis）  
・仮説を立て、その仮説が統計的に成立するといえるかを検証する  
・ただしなければ仮説を棄却し、反対の仮説を成立させる  
例）帰無仮説：薬Aと薬Bで効果の差はない＜ー統計的に差があると言えれば棄却する。言えなければ  採用する  
・「差がある」には様々なケースがある。無駄にケースがあり、全ての仮説をするには無理がある  
・「差がある」ことを言うために、「差がない」という１つの仮説を否定すればよい  
・対立仮説　ー＞（p１ ≠ p２：両側検定）、（p < p2, p1 > p2 :片側検定）  


## 検定の流れ  
１、帰無仮説と対立仮説を立てる  
２、帰無仮説が正しいと仮定の元、標本観察を行う（標本統計量を計算する）  
３、標本観察の結果、帰無仮説を否定できるかを確認する（否定した場合、対立仮説が成立）  
・２つの標本から比率を計算し、比率の差がどのような標本分布になるかを考えられる  
・帰無仮説が正しいとした場合、得られた比率の差がどれくらいの確率で得られるかを考える.   

## 優位水準  
・帰無仮説を棄却する境目のこと、帰無仮説を正しいとすると滅多におきない確率を優位水準という  
・通常はα＝.０５もしくはα=.01  
・検定統計量が優位水準以下の確率で得られる場合、帰無仮説を棄却する  
・帰無仮説が棄却された時「統計的に有意である」という.   

## p値
・帰無仮説が正しいとした時にその統計推定量がえられる確率をp値という  

## 検定結果パターン
・帰無仮説を棄却できた　ー＞　有意差あり　ー＞「差がある」と言える　ー＞対立仮説は成立する  
・帰無仮説を棄却できなかった　ー＞有意差なし　ー＞「差がない」ことを否定できない　ー＞対立仮説は成立しない  


## 比率差の検定(zテスト）
statsmodels.stats.proportion.proportions_ztest(count, nobs, alternative)
- count:　標本の中で実際に観察できた数
- nobs(the number of observations): 標本の大きさ
- alternative: 'two-sided', 'smaller', 'larger',  ('smaller':p1 < p2, 'larger':p1>p2)
- 母数に差があるのかを検定している　  
- 検定によって成立させたい仮説を対立仮説に選べばOK. 


```python
proportions_ztest([30, 33], [1000, 1000], alternative='smaller')
```
