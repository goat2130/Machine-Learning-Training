```python
import math

def uni_hist(N):
    x = np.random.uniform(0.0, 1.0, N)
    y = np.random.uniform(0.0, 1.0, N)

    plt.subplot(2, 1, 1)
    plt.hist(x)
    plt.title("No1: histgram")

    plt.subplot(2, 1, 2)
    plt.hist(y)
    plt.title("No2: histgram")

    plt.grid(True)
    plt.tight_layout()

uni_hist(10000)
```

```python
N = 1000
uni_hist(N)
```

```python
uni_hist(100)
```

```python
N = 10000

x = np.random.uniform(0.0, 1.0, N)
y = np.random.uniform(0.0, 1.0, N)

inside_x = []
inside_y = []

outside_x = []
outside_y = []

count_inside = 0
for count in range(0, N):
    d = math.hypot(x[count], y[count])
    if d < 1:
        count_inside += 1
        inside_x.append(x[count])
        inside_y.append(y[count])
    else:
        outside_x.append(x[count])
        outside_y.append(y[count])


print(count_inside)
```

```python
plt.figure(figsize=(5, 5))

circle_x = np.arange(0, 1, 0.001)
circle_y = np.sqrt(1 - circle_x * circle_x)

plt.plot(circle_x, circle_y)
plt.scatter(inside_x, inside_y, color="r")
plt.scatter(outside_x, outside_y, color="b")

plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
```

```python
print(4.0 * count_inside / N)
```
